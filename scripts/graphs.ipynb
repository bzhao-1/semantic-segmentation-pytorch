```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day = [.4763, .3422] 
night = [.0219, .3002]

# Plot the bars
plt.bar([1,3], day, color='gold', label='Tested on Day')
plt.bar([2,4], night, color='slateblue', label='Tested on Night')

# Add labels
plt.xlabel('Model', fontweight='bold')
plt.ylabel('MIOU', fontweight='bold')
plt.title('MIOU for models Trained on Day')
plt.xticks([1.5, 3.5], ['Camera Only (Base)', 'Lidar Only'])
plt.legend()
plt.grid(axis = 'y', linestyle = '--', linewidth = 0.5)
plt.ylim(0, .6)

# Show the plot
plt.show()
```

```python
day = [.4763, .3422, 0.4349, 0.4515] 
night = [.0219, .3002, 0.0767, 0.0592]

# Plot the bars with adjusted positions
plt.bar([1, 3, 6, 8], day, color='gold', label='Tested on Day')
plt.bar([2, 4, 7, 9], night, color='slateblue', label='Tested on Night')

# Add labels
plt.xlabel('Model', fontweight='bold')
plt.ylabel('MIOU', fontweight='bold')
plt.title('MIOU for models Trained on Day')
plt.xticks([1.5, 3.5, 6.5, 8.5], ['Camera Only (Base)', 'Lidar Only', 'Mid Fusion + C3', "Mid Fusion + C3 + Deepsup"], rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', linewidth=1, which='major')
plt.grid(axis='y', linestyle='--', linewidth=.3, which='minor')
plt.yticks(np.arange(0, 0.7, 0.1))
plt.minorticks_on()
plt.ylim(0, .6)
plt.tick_params(which='minor', width=0.2)

# Show the plot
plt.show()
```

```python
speeds = [30, 1.05, 0.73]
labels = ['Real Time', 'Base Model', 'Mid Fusion']
colors = ['gold', 'khaki', 'khaki']

plt.bar([1, 2, 3], speeds, color=colors)
plt.ylabel('Frames Per Second')
plt.xlabel('Model')
plt.title('Inference Speeds')
plt.ylim(0, 35)

plt.xticks([1, 2, 3], labels)

# Add values to bars
for i, v in enumerate(speeds):
  plt.text(i + 1, v + 0.5, str(v), ha='center')
```
